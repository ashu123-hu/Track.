```python
import random

width = 10
height = 10

score=0

snake = [(5, 5)]
direction = "RIGHT"

food = (random.randint(0, width-1), random.randint(0, height-1))

def print_board():
    for y in range(height):
        for x in range(width):
            if (x, y) in snake:
                print("S", end=" ")
            elif (x, y) == food:
                print("F", end=" ")
            else:
                print(".", end=" ")
        print()
    print()

while True:
    print_board()
    move = input("Move (W/A/S/D): ").upper()

    head_x, head_y = snake[0]

    if move == "W":
        head_y -= 1
    elif move == "S":
        head_y += 1
    elif move == "A":
        head_x -= 1
    elif move == "D":
        head_x += 1
    else:
        print("Invalid move!")
        continue

    new_head = (head_x, head_y)

    # Wall collision
    if head_x < 0 or head_x >= width or head_y < 0 or head_y >= height:
        print("Game Over! Hit the wall.")
        break

    # Self collision
    if new_head in snake:
        print("Game Over! Hit yourself.")
        break

    snake.insert(0, new_head)

    if new_head == food:
       score += 1
       print("Score:", score)
       food = (random.randint(0, width-1), random.randint(0, height-1))
    else:
        snake.pop()
```
